```python
### Large Averaging ####

topdir='large_250/'
wdir=glob.glob(topdir+'data_equil*/')
for j in range (len(wdir)):
    if j==0: 
        large_g1=np.genfromtxt(wdir[j]+'g1.dat')
        large_g3=np.genfromtxt(wdir[j]+'msd_coms.dat')
    else:
        temp_g1=np.genfromtxt(wdir[j]+'g1.dat')
        temp_g3=np.genfromtxt(wdir[j]+'msd_coms.dat')
        large_g1+=temp_g1
        large_g3+=temp_g3
large_g1=large_g1/j
large_g3=large_g3/j

plt.loglog(large_g1[:,0],large_g1[:,1],'k-',label='g1')
plt.loglog(large_g3[:,0],large_g3[:,1],label='g3')
plt.plot(large_g3[5000:24000,0],0.00015*large_g3[5000:24000,0]**(1),'k--')
#plt.plot(large_g3[0:3,0],0.07*large_g3[0:3,0]**(1/4),'k--')
plt.plot(large_g3[1:80,0],0.4*large_g3[1:80,0]**(1/2),'k--')
plt.show()
```

```python
### Small Averaging ####

topdir='small_96/'
wdir=glob.glob(topdir+'data_equil*/')
for j in range (len(wdir)):
    if j==0: 
        small_g1=np.genfromtxt(wdir[j]+'g1.dat')
        small_g3=np.genfromtxt(wdir[j]+'msd_coms.dat')
    else:
        temp_g1=np.genfromtxt(wdir[j]+'g1.dat')
        temp_g3=np.genfromtxt(wdir[j]+'msd_coms.dat')
        small_g1+=temp_g1
        small_g3+=temp_g3
small_g1=small_g1/j
small_g3=small_g3/j

plt.loglog(small_g1[:,0],small_g1[:,1],'k-',label='g1')
plt.loglog(small_g3[:,0],small_g3[:,1],label='g3')
plt.plot(small_g3[5000:24000,0],0.00015*small_g3[5000:24000,0]**1,'k--')
plt.plot(small_g3[0:3,0],0.07*small_g3[0:3,0]**(3/4),'k--')
plt.plot(small_g3[1:80,0],0.85*small_g3[1:80,0]**(1/2),'k--')
plt.show()
```

```python
### L250/L96 0.16/0.84 Averaging ####

topdir='L250_L96_0.16//'
wdir=glob.glob(topdir+'data_equil*/')
for j in range (len(wdir)):
    if j==0: 
        small_16_g1=np.genfromtxt(wdir[j]+'small_g1.dat')
        small_16_g3=np.genfromtxt(wdir[j]+'small_msd_coms.dat')
        large_16_g1=np.genfromtxt(wdir[j]+'large_g1.dat')
        large_16_g3=np.genfromtxt(wdir[j]+'large_msd_coms.dat')
    else:
        #temp_g1=np.genfromtxt(wdir[j]+'g1.dat')
        #temp_g3=np.genfromtxt(wdir[j]+'msd_coms.dat')
        temp_small_16_g1=np.genfromtxt(wdir[j]+'small_g1.dat')
        temp_small_16_g3=np.genfromtxt(wdir[j]+'small_msd_coms.dat')
        temp_large_16_g1=np.genfromtxt(wdir[j]+'large_g1.dat')
        temp_large_16_g3=np.genfromtxt(wdir[j]+'large_msd_coms.dat')
        small_16_g1+=temp_small_16_g1
        small_16_g3+=temp_small_16_g3
        large_16_g1+=temp_large_16_g1
        large_16_g3+=temp_large_16_g3
   
small_16_g1=small_16_g1/j
small_16_g3=small_16_g3/j
large_16_g1=large_16_g1/j
large_16_g3=large_16_g3/j

plt.loglog(small_16_g1[:,0],small_16_g1[:,1],'k-',label='g1_small')
plt.loglog(small_16_g3[:,0],small_16_g3[:,1],label='g3_small')
plt.loglog(large_16_g1[:,0],large_16_g1[:,1],'r-',label='g1_large')
plt.loglog(large_16_g3[:,0],large_16_g3[:,1],label='g3_large')
#plt.plot(small_g3[5000:24000,0],0.00015*small_g3[5000:24000,0]**1,'k--')
#plt.plot(small_g3[0:3,0],0.07*small_g3[0:3,0]**(3/4),'k--')
#plt.plot(small_g3[1:80,0],0.85*small_g3[1:80,0]**(1/2),'k--')
plt.legend(frameon=False)
plt.show()
```
